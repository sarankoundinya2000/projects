```python
def Knn_neigh(dataset,pre_set,k=3):
    for i in dataset:
        for j in dataset[i]:
            euclid_dist=np.sqrt(np.sum((np.array(j)-np.array(new_fe))**2))
            dist.append([euclid_dist,i])
    votes=[i[1] for i in sorted(dist)[:k]]
    most=Counter(votes).most_common(1)[0][0]
    
    return most
result=Knn_neigh(dataset,new_fe)
```
